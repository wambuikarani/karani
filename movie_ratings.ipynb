```python
import pandas as pd
user_item_details=pd.read_csv("./Datasets/movie_ratings.csv")
user_item_details
```

```python
import pandas as pd
movie_details=pd.read_csv("./Datasets/movie_titles.csv")
movie_details.head(10)
```

```python
# merge the two datasets
data=pd.merge(user_item_details,movie_details,on='movie_id')
data.head(10)
```

```python
data['movie_id'].sort_values(ascending=True)
```

```python
# best rated movies
data.groupby('title')['rating'].mean().sort_values(ascending=False).head(10)
```

```python
# get the number of rating per movie
mean_ratings=pd.DataFrame(data.groupby('title')['rating'].mean())
mean_ratings
mean_ratings['numb_of_ratings']=pd.DataFrame(data.groupby('title') ['rating'].count())
mean_ratings.head(10)
```

```python
mean_ratings.groupby('title')['numb_of_ratings'].mean().sort_values(ascending=False).head(10)
```
