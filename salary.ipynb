```python
import pandas as pd
data=pd.read_csv("./Datasets/Salary.csv")
data
```

```python
data.isnull() .sum()
```

```python
# splitting the data
array=data.values
x=array[:,0:2]
y=array[:,1]
array
```

```python
x=data[["YearsExperience"]]
y=data['Salary']
```

```python
# splitting our variables into training and testing tests  
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# cross validation
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression


# create an empty list and appenda all models ,give the models an alias name
models=[]
models.append(('D Trees',DecisionTreeRegressor()))
models.append(('KNC',KNeighborsRegressor()))
models.append(('Random forest',RandomForestRegressor()))
models.append(('Gradient boosting',GradientBoostingRegressor()))
models.append(('support Machines',SVR(gamma='auto')))
models.append(('LR',LinearRegression()))



```

```python
# import cros validation and Kfold
from sklearn.model_selection import cross_val_score,KFold
# create a foer loop to test each model in turn
for name ,model in models:
    kfold=KFold(n_splits=10,random_state=42,shuffle=True)
    cv_results=cross_val_score(model,x_train,y_train,cv=kfold,scoring='r2')
    print(name,'Outcome:=',cv_results.mean())
```

```python
model=GradientBoostingRegressor()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
print("predicted: ",predictions)
print("expected:",y_test)
```

```python
from sklearn.metrics import r2_score
print("Accuracy",r2_score(predictions,y_test))
```
