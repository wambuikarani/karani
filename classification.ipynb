```python
import pandas as pd
data=pd.read_csv ("./Datasets/recipe.csv")
data
```

```python
# check for empties
data.isnull().sum()
```

```python
# analyze data
data.describe()
```

```python
array=data.values
array

x=array[0:50,0:8]
y=array[0:50,8]
x
y

```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
# passing our trained raws into an algorithm :support vector classifier
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
# test our algorithm
prediction=model.predict(x_test)
print(f"The  algo predicted: {prediction}")
print(f"The actual anwers are :{y_test}")
```

```python
# test the accuracy score
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(prediction,y_test)*100
accuracy
```

```python
# use algo to predict what we are making
ingredients=[[5,20,10,0,15,8,10,11]]
cake=model.predict(ingredients)
cake
```

```python
# import all the clasification models
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC


# create an empty list and appenda all models ,give the models an alias name
models=[]
models.append(('D Trees',DecisionTreeClassifier()))
models.append(('Gaussian',GaussianNB()))
models.append(('KNC',KNeighborsClassifier()))
models.append(('Random forest',RandomForestClassifier()))
models.append(('Gradient classifier',GradientBoostingClassifier()))
models.append(('svm',SVC()))
models




```

```python
from sklearn.model_selection import cross_val_score,KFold

# create a for loop so as to test each model in array
for name,model in models:
    # we do a 10 split kfold
    kfold=KFold(n_splits=10,random_state=42,shuffle=True) 
    # Get the resultys for each fold
    cv_results=cross_val_score(model,x_train,y_train,cv=kfold,scoring="accuracy")
    # average of all folds
    print(f'Results:={model}-{cv_results.mean()}')


```
