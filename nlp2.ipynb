```python
# NLP uses unstructured data
import requests   #This allows you to send HTTP requests in python
from  bs4 import BeautifulSoup #bs4 is a python libraray for pulling data out of HTML and XML files.
url="https://www.standardmedia.co.ke/sports/football/article/2001495620/pochettino-leaves-chelsea-after-just-one-season-in-charge"
page = requests.get(url)
soup= BeautifulSoup(page.content,"html.parser")
div= soup.findAll("div",class_="story")
# soup=BeautifulSoup(str(div))#beautiful oup is used to extract data from web pages

text=soup.get_text()
print(text)
```

```python
nltk.download("punkt")
from nltk.tokenize import sent_tokenize,word_tokenize
tokenized_sent=sent_tokenize(text)
tokenized_word=word_tokenize(text)

print(tokenized_sent)
print(tokenized_word)
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
print(stopwords.fileids())

# get English stop words
mystopwords=set(stopwords.words('english'))
print(mystopwords)
```

```python
filtered_words=[]
for word in tokenized_word:
    if word.lower()not in  mystopwords:
        filtered_words.append(word)
print(f"old words:{tokenized_word}")
print(f"old words:{filtered_words}")
```

```python
from wordcloud import wordcloud,STOPWORDS
from PIL import Image
# Function to create wordcloud
def create_wordcloud(text):
    stopwords=set(STOPWORDS)
    wc=wordcloud(background_color="white",
    max_words=3000,
    stopwords=stopwords,repeat=True)
    wc.generate(str(text))
    wc.to_file("wc.png")
    print("word cloud saved successfully")
    path="wc.png"
    display(Image.open(path))
```

```python
# split to training and testing test
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(data["text"],data['airline_sentiment'],test_size=0.3,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer #used tp represent text in numericals
```

```python
# remove stopwordsfrom xtest and ytest then convert into numericals representation suitable for machine learning task using scikits count vectorizer
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(lowercase=True,stop_words='english')
x_train_new=vectorizer.fit_transform(x_train)
x_train_new
```

```python
# create and train a logistic regression model for classification
from sklearn.linear_model import LogisticRegression,SGDClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

model=LogisticRegression()
model.fit(x_train_new,y_train)
```

```python
# test model
from sklearn.metrics import accuracy_score 
x_test_new=vectorizer.transform(x_test)
predictions=model.predict(x_test_new)
print(data.groupby("airline_sentiment").size())
```

```python
print(f"Accuracy:{accuracy_score(y_test,predictions)}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
