```python
import nltk
from nltk.tokenize import PunktSentenceTokenizer
```

```python
# NLP uses unstructured data
import requests   #This allows you to send HTTP requests in python
from  bs4 import BeautifulSoup #bs4 is a python libraray for pulling data out of HTML and XML files.
url="https://www.standardmedia.co.ke/sports/football/article/2001495620/pochettino-leaves-chelsea-after-just-one-season-in-charge"
page = requests.get(url)
soup= BeautifulSoup(page.content,"html.parser")
div= soup.findAll("div",class_="story")
# soup=BeautifulSoup(str(div))#beautiful oup is used to extract data from web pages

text=soup.get_text()
print(text)
```

```python
nltk.download("punkt")
from nltk.tokenize import sent_tokenize,word_tokenize
tokenized_sent=sent_tokenize(text)
tokenized_word=word_tokenize(text)

print(tokenized_sent)
print(tokenized_word)
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
print(stopwords.fileids())

# get English stop words
mystopwords=set(stopwords.words('english'))
print(mystopwords)
```

```python
filtered_words=[]
for word in tokenized_word:
    if word.lower()not in  mystopwords:
        filtered_words.append(word)
print(f"old words:{tokenized_word}")
print(f"old words:{filtered_words}")
```
