```python
import pandas as pd
data=pd.read_csv("./Datasets/cleanedSchool.csv")
data
```

```python
import scipy.stats as stats
```

<!-- perform the one sample test -->
stats.test_1samp(a=data['English'],popmean=60)

```python
stats.ttest_1samp(a=data['English'],popmean=60)

t_statistic, p_value=stats.ttest_1samp(a=data['English'],popmean=60)

# print the results
print(f"T_Statistic:{t_statistic}")
print(f"p_Value: {p_value}")

if p_value>0.05:
    print("There is sufficient evidence to conclude that the mean English score is significantly different from 50")
else:
    print("There is insufficient evidence to conclude that the mean English score is different from 50")
```
