```python
# cleaning.
# do we have empties
# do we have unwanted columns or rows?

data.isnull().sum()
```

```python
# filL empties
data['Gender'].fillna(2.0,inplace=True)
data['Rank'].fillna(0.0,inplace=True)
data['Major'].fillna(0.0,inplace=True)
data['Smoking'].fillna(3.0,inplace=True)
data['LiveOnCampus'].fillna(1.0,inplace=True)
data['bday'].fillna(1.0,inplace=True)
data['enrolldate'].fillna(1.0,inplace=True)
data['expgradate'].fillna(1.0,inplace=True)
data['Height'].fillna(1.0,inplace=True)
data['Weight'].fillna(1.0,inplace=True)
data['Sprint'].fillna(1.0,inplace=True)
data['English'].fillna(1.0,inplace=True)
data['Reading'].fillna(1.0,inplace=True)
data['Math'].fillna(1.0,inplace=True)
data['Writing'].fillna(1.0,inplace=True)
data['State'].fillna(1.0,inplace=True)
data['HowCommute'].fillna(1.0,inplace=True)
data['CommuteTime'].fillna(1.0,inplace=True)
data['SleepTime'].fillna(1.0,inplace=True)
data['StudyTime'].fillna(1.0,inplace=True)
data['MileMinDur'].fillna(1.0,inplace=True)















```

```python
data.isnull().sum()
```

```python
# replacing
data['Gender'].replace({
    0:'Male',
    1:'Female',
    2:'Other'
},inplace=True
)

data['LiveOnCampus'].replace(
    {
        0:'off-campus',
        1:'on-campus'
    },inplace=True
)

data['Athlete'].replace(
    {
        0:'Non-athlete',
        1:'Athlete'
    },inplace=True
)

data['Smoking'].replace(
    {
        0:'Nonsmoker',
        1:'pastsmoker',
        2:'Currentsmoker'
    },inplace=True
)

data['HowCommute'].replace(
    {
        1:'Walk',
        2:'Bike',
        3:'Car',
        4:'Public transit',
        5:'Other'
    },inplace=True
)
```

```python
data.head(10)
```

```python
#function for converting weight from pounds to kgs(1kg=2.2pounds)
def convert(x):
    ans=x/2.20462
    return ans
data['Weight']=data['Weight'].apply(convert)
data['Weight'].head(5)
```

```python
# converting height from inches to meter(1inch=0.0254meter)
def converts (y):
    answer=y*0.0254
    return answer
data['Height']=data['Height'].apply (converts)
data['Height'].head(5)
```

```python
data.to_csv('./Datasets/cleanedSchool.csv')
```
